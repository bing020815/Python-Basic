## Groupby

* How Groupby Works?
    + Split a DataFrame into groups based on some criteria
    + Apply a function to each group independently
    + Combine the results into a DataFrame

* groupby()
    + pnadas.DataFrame.groupby(by=None, axis=0, level=None, as_index=True, sort=True, group_keys=True, squeeze=False, **kwargs)
    + Returns a groupby object

```python
oo.groupby('Edition')
```

```python
type(oo.groupby('Edition'))
```

```python
list(oo.groupby('Edition'))
```

* Iterate through a Group
    + for key, group in DataFrame.groupby():
            print(key)
            print(group)

```python
for group_key, group_value in oo.groupby('Edition'):
    print(group_key)
    print(group_value)
```

```python
type(group_value) # group_value is not 2008 group
```

## groupby computation

* Groupby Computation:
    + GroupBy.size()
    + GroupBy.count()
    + GroupBy.first(), GroupBy.last()
    + GroupBy.head(), GroupBy.tail()
    + GroupBy.mean()
    + GroupBy.max(), GroupBy.min()

On the Job:
    + agg() - multilple statistics in one calculation per group
    + DataFrame.groupby(agg([...]))
    + DataFrame.groupby(agg({..:[...]}))

```python
oo.groupby('Edition').size()
```

### agg([...])

```python
oo.groupby(['Edition', 'NOC','Medal']).agg(['min','max', 'count'])
```

```python
oo.groupby(['Edition', 'NOC','Medal']).agg('count')
```

```python
# Get a cleaner output by using size()
oo.groupby(['Edition', 'NOC','Medal']).size()
```

### agg({:[...]})

```python
oo.groupby(['Edition', 'NOC','Medal']).agg({'Edition':['min','max', 'count']})
```

### Challenge:

### In every Olympics, which US athlete has won the most total number of medals? Include the athlete's discipline.

```python
gy = oo[oo.NOC=='USA']
gy = gy.groupby(['Edition','Athlete', 'Medal']).size().unstack('Medal', fill_value=0)
gy['Total'] =  gy['Gold'] + gy['Silver'] + gy['Bronze'] # Create a new column of total
gy = gy[['Gold', 'Silver','Bronze', 'Total']]
gy.reset_index(inplace=True)
# store the top usa athlete dataframe for each group(Edition) into a list
tu = [group.sort_values('Total', ascending=False)[:1] for year, group in gy.groupby('Edition')] 
tu # a list of dataframes
# by storing the result into a list, then, we can populate the dataframe from the elements in the list
top = pd.DataFrame() # create a new empty dataframe
for i in tu:
    top = top.append(i)
top
```

```python
gy = oo[oo.NOC=='USA']
gy = gy.groupby(['Edition','Athlete', 'Medal']).size().unstack('Medal', fill_value=0)
gy['Total'] =  gy['Gold'] + gy['Silver'] + gy['Bronze'] # Create a new column of total
gy = gy[['Gold', 'Silver','Bronze', 'Total']]
gy.reset_index(inplace=True)
del gy.columns.name # delete the columns index name
gy
# store the top usa athlete dataframe for each group(Edition) into a list
tu = [group.sort_values('Total', ascending=False)[:1] for year, group in gy.groupby('Edition')] 
tu # a list of dataframes
# by storing the result into a list, then, we can populate the dataframe from the elements in the list
top = pd.DataFrame() # create a new empty dataframe
for i in tu:
    top = top.append(i)
top.reset_index(inplace=True,)
top.drop(['index'], axis=1)
```
